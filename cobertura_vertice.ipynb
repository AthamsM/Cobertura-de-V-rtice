```python
import networkx as nx

def ler_data(Data):
    grafo = nx.Graph()
    with open(Data, 'r') as f:
        for line in f:
            u, v = map(int, line.split())  # Converte os vértices para inteiros
            grafo.add_edge(u, v)  # Adiciona a aresta ao grafo
    return grafo
```

```python
def cobertura_vertice_conjunto(grafo):
    cobertura_vertice = set()
    arestas = []

    for u in grafo:
        for v in grafo[u]:
            if (u, v) not in arestas and (v, u) not in arestas:
                arestas.append((u, v))
    for u, v in arestas:
        if u not in cobertura_vertice and v not in cobertura_vertice:
            cobertura_vertice.add(u)
            cobertura_vertice.add(v)

    return cobertura_vertice
```

```python
def cobertura_vertice_gulosa(grafo):


    arestas_cobertas = set()
    cobertura_vertice = set()

    arestas = []
    for u in grafo:
        for v in grafo[u]:
            if (u, v) not in arestas and (v, u) not in arestas:
                arestas.append((u, v))


    while len(arestas_cobertas) < len(arestas):

        # Dicionário para contar quantas arestas cada vértice pode cobrir
        vertice_aresta_count = {}
        for vertice in grafo:
            count = 0
            for aresta in arestas:
                if vertice in aresta and aresta not in arestas_cobertas:
                    count += 1
            vertice_aresta_count[vertice] = count

        melhor_vertice = None
        maior_cobertura = -1
        for vertice, count in vertice_aresta_count.items():
            if count > maior_cobertura:
                maior_cobertura = count
                melhor_vertice = vertice

        cobertura_vertice.add(melhor_vertice)

        # Atualizar o conjunto de arestas cobertas
        for aresta in arestas:
            if melhor_vertice in aresta:
                arestas_cobertas.add(aresta)


    return cobertura_vertice
```

```python
if __name__ == "__main__":

    Data = "large_large_graph.txt"
    grafo = ler_data(Data)
```

```python
vertex_cover = cobertura_vertice_gulosa(grafo)

print("Cobertura de Vértices:", vertex_cover)
```

```python
cover_vertex = cobertura_vertice_conjunto(grafo)

print("Cobertura de Vértices:", cover_vertex)
```

```python
import matplotlib.pyplot as plt

def read_graph_from_txt(file_path):
    """
    Lê um arquivo de texto contendo arestas e cria um grafo NetworkX.
    """
    graph = nx.Graph()

    with open(file_path, 'r') as file:
        for line in file:
            # Cada linha contém dois números representando uma aresta
            u, v = map(int, line.split())
            graph.add_edge(u, v)

    return graph

def plot_graph(graph, title="Grafo"):
    """
    Desenha o grafo usando NetworkX e Matplotlib.
    """
    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(graph)  # Layout para organizar os vértices
    nx.draw(
        graph, pos,
        with_labels=True,
        node_color='lightblue',
        edge_color='gray',
        node_size=500,
        font_size=10
    )
    plt.title(title)
    plt.show()

# Caminho do arquivo de texto
file_path = "large_graph.txt"

# Ler o grafo do arquivo
graph = read_graph_from_txt(file_path)

# Desenhar o grafo
plot_graph(graph, title="Visualização do Grafo")
```
